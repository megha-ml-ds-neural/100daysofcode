```
# MULTILINE PLOTS
```

```
x=range(1,5)
plt.plot(x,[x**1.5 for x in x])
plt.plot(x,[x**3 for x in x])
plt.plot(x,[x**4.5 for x in x])
plt.show()
```

plot() supports another syntax useful in this situation. We can plot multiline figures by passing the X and Y values list in a single plot() call:


```
x=range(1,5)
plt.plot(x,[x**2 for x in x],x,[x**4 for x in x],x ,[x**6 for x in x ])
plt.show()
```

With the usage of powerful tool arange of numpy()

```
 x = np.arange(1, 5) 
plt.plot(x,x**1.5,x**3.0,x/3.0)
plt.show() 
```

ADDING A GRID

```
x=np.arange(1,5)
plt.plot(x,x**1.5,x,x**3.0)
plt.grid(True)
plt.show()
```

HANDLING AXES


```
x=np.arange(1,5)
plt.plot(x,x**1.5,x,x**3.0,x,x/3.0)
plt.axis()
```

```
x=np.arange(1,5)
plt.plot(x,x**1.5,x,x**3.0,x,x/3.0)
plt.axis([0,5,-1,13])
plt.show()
```

```
x=np.arange(1,5)
plt.plot(x,x**1.5,x,x**3.0,x,x/3.0)
plt.xlim(1.0,4,0)
plt.ylim(0.0,12.0)

# these are used to limit the x and y axis
```

ADDING LABELS

```
y=np.arange(1,6,0.01)
plt.plot(y)
plt.grid(True)
plt.xlabel("This is x label")
plt.ylabel("This is ylabel")
plt.show()
```

TITLE AND LEGENDS

```
plt.plot([1,2,3,4,5])
plt.title("Simple Plot")
plt.show()
```

ADDING A LEGEND

```
m=np.arange(1,6,0.01)
plt.plot(m,m**1,label='Normal')
plt.plot(m,m**2,label='Quite fast')
plt.plot(m,m**3,label='faster')
plt.title("graph")
plt.grid()
plt.show()
```

***A COMPLETE EXAMPLE***

```
import matplotlib.pyplot as plt
import numpy as np
x=np.arange(1,5)
plt.plot(x,x*1.5,label='Normal')
plt.plot(x,x*2.5,label='Quite Normal')
plt.plot(x,x*3.5,label='Fast Normal')
plt.plot(x,x*4.5,label='Fater Normal')
plt.grid(True)
plt.xlabel("Samples")
plt.ylabel("Values Measured")
plt.title("Sample growth")
plt.show()
```
